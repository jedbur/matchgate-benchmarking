```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
```

## Matchgate Benchmarking - Plots

Below is the code used to produce plots in our [paper](https://arxiv.org/abs/2404.07974). The data from our simulations is stored in the `data_mgc.csv` and `data_givens.csv` files. Similar data may be obtained by running our simulations in the `matchgate_benchmarking_data_collction.ipynb` notebook.

```python
#%matplotlib ipympl

# Read the CSV file into a DataFrame
df = pd.read_csv('data_mgc.csv', header=None, names=['A', 'B', 'C'])

# Ensure that 'a', 'b', and 'c' are numeric
df['A'] = pd.to_numeric(df['A'], errors='coerce')
df['B'] = pd.to_numeric(df['B'], errors='coerce')
df['C'] = pd.to_numeric(df['C'], errors='coerce')


# Create a scatter plot
plt.scatter(df['A'], df['B'], c=df['C'], cmap='viridis', edgecolor='k', alpha=0.75)

# Set the color bar to show the correspondence of 'c' values to colors
plt.colorbar(fraction=0.046, pad=0.04).ax.tick_params(labelsize=9)

# Add the dashed line for a=b
max_value = max(df['A'].max(), df['B'].max())
plt.plot([0.9, 1.0], [0.9, 1.0], 'k--', label='a = b', alpha=0.5)

# Set the labels for the axes
plt.xlabel('Estimated Entanglement Fidelity', fontsize='small')
plt.ylabel('True Entanglement Fidelity', fontsize='small')

# Add title
plt.title('Random MG Circuit; Depolarising Noise; $n=3$, $\epsilon=0.025$, $\delta = 0.05$', fontsize='medium')

# Show the plot
plt.show()
```
